## Ejercicio 5: 

#### Diseñar un procedimiento feigen que dibuje el diagrama de Feigenbaum de una familia uniparamétrica de funciones, para valores del parametro en cierto intervalo acotado. El input debe ser la familia uniparamétrica de funciones fc(x), los extremos a < b del intervalo de parámetros, y el punto x0 del cual se calcula la órbita; el output debe ser una representación del diagrama de Feigenbaum.

```python
import numpy as np
import matplotlib.pyplot as plt

def feigen(f, a, b, *, x0=0):
    n_iter = 1500
    n_keep = 200
    n_r = 20000
    figsize = (8,6)
    
    plt.figure(figsize=figsize)
    rs   = np.linspace(a, b, n_r)
    xn   = np.full(n_r, x0, dtype=np.float64)

    for _ in range(n_iter):
        xn = f(xn, rs)
        if _ >= n_iter - n_keep:
            plt.scatter(rs, xn, s=0.02, color='b')
            
    plt.show()
```

## Ejercicio 7:

```python
def f_cos(x, c):
    return np.cos(c * x)
```

```python
feigen(f_cos, a = 1.35, b = 1.6, x0 = 0.5)

feigen(f_cos, a = 2.75, b = 3.27, x0 = 0.5)

feigen(f_cos, a = 4.55, b = 4.75, x0 = 0.5)
```

```python
c_val = 2.9
def g(x): return f_cos(x, c_val)

# órbita desde varios x0
x0s = [0.5, 0.1, 1.2]
for x0 in x0s:
    x = x0
    print(f"\nórbita desde x0={x0}:")
    for k in range(15):
        x = g(x)
        print(f"{k+1:2d}  {x: .6f}")
```
